```python
import requests
from bs4 import BeautifulSoup
import time 


hiragana_list = ['あ', 'い', 'う', 'え', 'お', 'か', 'き', 'く', 'け', 'こ', 'さ', 'し', 'す', 'せ', 'そ', 
                 'た', 'ち', 'つ', 'て', 'と', 'な', 'に', 'ぬ', 'ね', 'の', 'は', 'ひ', 'ふ', 'へ', 'ほ', 
                 'ま', 'み', 'む', 'め', 'も', 'や', 'ゆ', 'よ', 'ら', 'り', 'る', 'れ', 'ろ', 'わ', 'を', 'ん']

# スクレイピング対象のページURL
base_url = 'https://keytube.net/artist/?initial=' 
response = requests.get(base_url + hiragana_list[0])
soup = BeautifulSoup(response.text, 'html.parser')


a_tags = soup.find_all('a')

# aタグのリストを表示
count = 0
for a in a_tags:
    time.sleep(1)
    link = a.get('href').split('/')
    # artistかどうか調べる
    if len(link) > 3 and 'artist' == link[-3]:
        
        count += 1
print(count)        
```
